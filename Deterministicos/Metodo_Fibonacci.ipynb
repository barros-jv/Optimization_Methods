## Métodos de Otimização: Métodos de procura 1-D

### Método de Fibonacci

__Imports:__

```python
import numpy as np
import matplotlib.pyplot as plt
```

__Função Avaliada:__

```python
def f(x):
    return -7-17*x+2*x**2

x = np.linspace(0, 8)

plt.plot(x,f(x))
plt.show()
```

__Método:__

```python
n = 5;
F = np.zeros( (n+1),dtype='float' )
I = np.zeros( (n+1),dtype='float' )
I[0]=x[-1]-x[0]
F[0] = 1; F[1] = 1

for i in range(2,n+1):
    F[i]=F[i-2]+F[i-1]
    
print(F)
```

```python
for i in range(0,n):
    F[n-2-i]/F[n-i]*I[i]

    if f(x[0])>f(x[0]+F[n-2-i]/F[n-i]*I[i]):
        Xi=x[0]+F[n-2-i]/F[n-i]*I[i]
    else:
        Xi=x[0]

    if f(x[-1])>f(x[-1]-F[n-2-i]/F[n-i]*I[i]):
        Xf=x[-1]-F[n-2-i]/F[n-i]*I[i]
    else:
        Xf=x[-1]

    x = np.linspace(Xi, Xf)
    I[i+1]=x[-1]-x[0]

layout = "{0:.5f}"
print(layout.format(x[0])+" "+layout.format(f(x[0])))
```
