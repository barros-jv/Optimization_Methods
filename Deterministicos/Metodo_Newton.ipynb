## Métodos de Otimização: Métodos de procura multidimensional

### Steepest Descent

__Imports:__

```python
from mpl_toolkits import mplot3d
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

__Função Avaliada:__

```python
def f(x, y):
    return 0.25*x**4-3.*x**3+11.*x**2-13.*x+0.25*y**4-3.*y**3+11.*y**2-13.*y

x1 = np.linspace(0, 6)
x2 = np.linspace(0, 6)

X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)
```

```python
plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.plot_surface(X1, X2, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

__Método:__

```python
def grad(x, y):
    G=np.zeros(2, dtype='float')
    G[0] = x**3 -9.*x**2 +22.*x - 13.
    G[1] = y**3 -9.*y**2 +22.*y - 13.
    return G
```

```python
Hinv=np.zeros((2,2), dtype='float')
G=np.zeros(2, dtype='float')
xnew=np.zeros(2, dtype='float')
xold=np.zeros(2, dtype='float')
kmax=20; k=0
xold[0]=5; xold[1]=5; F0 = f(xold[0],xold[1])
layout = "{0:.5f}"

while k < kmax:
    k += 1
    # Calculo Gradiente
    G = grad(xold[0],xold[1])
    #print(xold)
    #print(G)
    
    # calcula a inversa do Hessiano
    Hinv[0,0]=1./(3.*xold[0]**2 -18.*xold[0] +22.)
    Hinv[1,1]=1./(3.*xold[1]**2 -18.*xold[1] +22.)
    Hinv[0,1]=0.
    Hinv[1,0]=0.
    #print(Hinv)
    
    # Processo iterativo
    xnew=xold-np.matmul(Hinv, G)
    #print(np.dot(Hinv, G))
    #print(xnew)
    
    #Atualiza xold
    xold = xnew
    
    #Calculo da função
    F = f(xnew[0],xnew[1])
    print(layout.format(xnew[0])+" "+layout.format(xnew[1])+" "+layout.format(F))
    #print(layout.format(x1new,x2new,F))
```

```python
plt.figure(figsize=(12,12))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X1, X2, Z, color='black')

# Data for a three-dimensional line
zline = np.linspace(0, F)
xline = np.linspace(xold[0], xnew[0])
yline = np.linspace(xold[1], xnew[1])
ax.plot3D(xline, yline, zline, 'red', linewidth = 3)

plt.show()
```
